```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
imgreal = cv2.imread('Elon_Musk.jpg')
graypic = cv2.cvtColor(imgreal,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(graypic, cmap='gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(graypic, scaleFactor = 1.2, minNeighbors = 5);
```

```python
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
    cv2.rectangle(imgreal,(x,y), (x+w,y+h), (0,255,0),2)
```

```python
def convertToRGB(imageconvert):
    return cv2.cvtColor(imageconvert,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(convertToRGB(imgreal))
```
